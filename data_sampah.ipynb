```python
import pandas as pd

# No.2
data = [
    {'Kabupaten/Kota': 'Bandung', 'Produksi Sampah (ton)': 1200, 'Tahun': 2022},
    {'Kabupaten/Kota': 'Cimahi', 'Produksi Sampah (ton)': 800, 'Tahun': 2022},
    {'Kabupaten/Kota': 'Bekasi', 'Produksi Sampah (ton)': 1500, 'Tahun': 2022},
    {'Kabupaten/Kota': 'Bandung', 'Produksi Sampah (ton)': 1300, 'Tahun': 2023},
    {'Kabupaten/Kota': 'Cimahi', 'Produksi Sampah (ton)': 900, 'Tahun': 2023}
]

df = pd.DataFrame(data)

tahun_tertentu = 2023

total_produksi = 0
for index, row in df.iterrows():
    if row['Tahun'] == tahun_tertentu:
        total_produksi += row['Produksi Sampah (ton)']

print(f"Total produksi sampah di Jawa Barat pada tahun {tahun_tertentu}: {total_produksi} ton")

hasil_akhir = pd.DataFrame({
    'Tahun': [tahun_tertentu],
    'Total Produksi Sampah (ton)': [total_produksi]
})

hasil_akhir.to_csv('hasil_akhir_total_produksi.csv', index=False)
print("Data hasil akhir berhasil diekspor ke file CSV: hasil_akhir_total_produksi.csv")

hasil_akhir.to_excel('hasil_akhir_total_produksi.xlsx', index=False, engine='openpyxl')
print("Data hasil akhir berhasil diekspor ke file Excel: hasil_akhir_total_produksi.xlsx")
```

```python
import pandas as pd

# no.3
data = {
    'Kabupaten/Kota': ['Bandung', 'Cimahi', 'Bekasi', 'Bandung', 'Cimahi'],
    'Produksi Sampah (ton)': [1200, 800, 1500, 1300, 900],
    'Tahun': [2022, 2022, 2022, 2023, 2023]
}

df = pd.DataFrame(data)

produksi_per_tahun = {}

for index, row in df.iterrows():
    tahun = row['Tahun']
    produksi = row['Produksi Sampah (ton)']
    
    if tahun in produksi_per_tahun:
        produksi_per_tahun[tahun] += produksi
    else:
        produksi_per_tahun[tahun] = produksi

hasil_tabel = []

for tahun, total in produksi_per_tahun.items():
    hasil_tabel.append({'Tahun': tahun, 'Total Produksi (ton)': total})

df_hasil = pd.DataFrame(hasil_tabel)

("Jumlah produksi sampah per tahun:")
(df_hasil)
```

```python
import pandas as pd

#no.4
data = {
    'Kabupaten/Kota': ['Bandung', 'Cimahi', 'Bekasi', 'Bandung', 'Cimahi'],
    'Produksi Sampah (ton)': [1200, 800, 1500, 1300, 900],
    'Tahun': [2022, 2022, 2022, 2023, 2023]
}

df = pd.DataFrame(data)

produksi_per_kota_tahun = {}

for index, row in df.iterrows():
    kota = row['Kabupaten/Kota']
    tahun = row['Tahun']
    produksi = row['Produksi Sampah (ton)']
    
    if kota not in produksi_per_kota_tahun:
        produksi_per_kota_tahun[kota] = {}
    
    if tahun in produksi_per_kota_tahun[kota]:
        produksi_per_kota_tahun[kota][tahun] += produksi
    else:
        produksi_per_kota_tahun[kota][tahun] = produksi

hasil_tabel = []

for kota, tahun_data in produksi_per_kota_tahun.items():
    for tahun, total in tahun_data.items():
        hasil_tabel.append({'Kabupaten/Kota': kota, 'Tahun': tahun, 'Total Produksi (ton)': total})

df_hasil = pd.DataFrame(hasil_tabel)

("Jumlah produksi sampah per Kota/Kabupaten per tahun:")
(df_hasil)
```
